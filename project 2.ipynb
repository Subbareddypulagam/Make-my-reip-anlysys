```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import requests
from bs4 import BeautifulSoup

import re
```

```python
Hotelname_list=[]
list_urls=['goa','mumbai','mahabaleshwar','matheran','delhi','shimla','lansdowne','digha','bangalore','ooty','pune','jaipur','kolkatha','ahmedabad','darjeeling','manali','jaipur','mussoorie','udaipur','puri','mount_abu','lonavala','kodaikanal','shirdi','kodaikanal','pondicherry']
place_list=[]
n=0
for i in list_urls:
    URL='https://www.makemytrip.com/hotels/'+i+'-hotels.html'
    #print(URL)
    page=requests.get(URL,headers = {'User-Agent':'Mozilla/5.0'})
    pagecontent=page.text
    soup=BeautifulSoup(pagecontent)
    list_3=soup.find_all('div', class_="flexOne makeFlex")
    
    for j in list_3:
        Hotelname=j.find('span', class_="wordBreak appendRight10")
        place_list=place_list+[i]
        if Hotelname is None:
            Hotelname_list.append(np.NaN)
        else:
            Hotelname_list.append(Hotelname.text)
            
```

```python
price_list=[]
list_urls=['goa','mumbai','mahabaleshwar','matheran','delhi','shimla','lansdowne','digha','bangalore','ooty','pune','jaipur','kolkatha','ahmedabad','darjeeling','manali','jaipur','mussoorie','udaipur','puri','mount_abu','lonavala','kodaikanal','shirdi','kodaikanal','pondicherry']


for i in list_urls:
    URL='https://www.makemytrip.com/hotels/'+i+'-hotels.html'
    #print(URL)
    page=requests.get(URL,headers = {'User-Agent':'Mozilla/5.0'})
    pagecontent=page.text
    soup=BeautifulSoup(pagecontent)
    list_3=soup.find_all('div', class_="flexOne makeFlex")
    
    for j in list_3:
        price=j.find('p',attrs={'class':'latoBlack font26 blackText appendBottom5','id':'hlistpg_hotel_shown_price'})
        
        if price is None:
            price_list.append(np.NaN)
            
        else:
            price_list.append(price.text)
```

```python
rating_list=[]
list_urls=['goa','mumbai','mahabaleshwar','matheran','delhi','shimla','lansdowne','digha','bangalore','ooty','pune','jaipur','kolkatha','ahmedabad','darjeeling','manali','jaipur','mussoorie','udaipur','puri','mount_abu','lonavala','kodaikanal','shirdi','kodaikanal','pondicherry']

Ratings1=[]
Performance=[]

for i in list_urls:
    URL='https://www.makemytrip.com/hotels/'+i+'-hotels.html'
    #print(URL)
    page=requests.get(URL,headers = {'User-Agent':'Mozilla/5.0'})
    pagecontent=page.text
    soup=BeautifulSoup(pagecontent)
    list_3=soup.find_all('div', class_="flexOne makeFlex")
    
    for j in list_3:
        rating=j.find('div',attrs={'class':'mmtRatingTooltipCont'})
        rating.text.split('\n')
        ZZ=[]
        for i in rating.text.split('\n'):
            if len(i)<1:
                pass
            else:
                ZZ=ZZ+[i]
        Ratings1=Ratings1+[ZZ[0]]
        Performance=Performance+[ZZ[1]]
```

```python
list_urls=['goa','mumbai','mahabaleshwar','matheran','delhi','shimla','lansdowne','digha','bangalore','ooty','pune','jaipur','kolkatha','ahmedabad','darjeeling','manali','jaipur','mussoorie','udaipur','puri','mount_abu','lonavala','kodaikanal','shirdi','kodaikanal','pondicherry']

specialities_list=[]



for i in list_urls:
    URL='https://www.makemytrip.com/hotels/'+i+'-hotels.html'
    #print(URL)
    page=requests.get(URL,headers = {'User-Agent':'Mozilla/5.0'})
    pagecontent=page.text
    soup=BeautifulSoup(pagecontent)
    list_3=soup.find_all('div', class_="flexOne makeFlex")
    
    for j in list_3:
        specialities=j.find('div',attrs={'class':'flexOne appendLeft20'})
        
        if specialities is None:
            specialities_list.append(np.NaN)
        else:
            specialities_list.append(specialities.text)
```

```python
len(specialities_list)
```

```python
Location1_list=[]
for i in specialities_list:
    Location1=re.findall("Location[0-5].[0-9]",str(i))
        
    if Location1 is None:
        Location1_list.append(np.NaN)
    else:
        Location1_list.append(''.join(Location1))
```

```python
a=[]
for i in Location1_list:
    if len(i)<1:
        a.append(np.nan)
    else:
        a.append(float(i.replace("Location","")))      
```

```python
friendliness1_list=[]
for i in specialities_list:
    friendliness1=re.findall('([friendliness]{12}[0-5].[0-9])',str(i))
        
    if friendliness1 is None:
        friendliness1_list.append(np.NaN)
    else:
        friendliness1_list.append(''.join(friendliness1))
```

```python
b=[]
for i in friendliness1_list:
    if len(i)<1:
        b.append(np.nan)
    else:
        b.append(float(i.replace("friendliness",""))) 
```
